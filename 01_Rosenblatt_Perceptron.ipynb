# Question 1 - McCulloch-Pitts Neuron

```python
# Define the neuron
def MCP(x, w):
    return np.sign(np.dot(w, x))
```

```python
w = np.array([3, 2, 2])
x_1 = np.array([-1, -1, -1])
x_2 = np.array([-1, 1, -1])
x_3 = np.array([-1, -1, 1])
x_4 = np.array([-1, 1, 1])

print("McCulloch-Pitts Neuron performs logical AND operator:")

print("Case 1 - False and False: {}".format(MCP(x_1, w)))
print("Case 2 -True and False: {}".format(MCP(x_2, w)))
print("Case 3 -False and True: {}".format(MCP(x_3, w)))
print("Case 4 -True and True: {}".format(MCP(x_4, w)))
```

# Question 2 - Activation Functions

```python
# Define the required functions - Sigmoid, Tanh, Piecewise Linear Unit
def sigmoid(x, a=1):
    return 2/(1+np.exp(-a*x)) - 1

def tanh(x, a=1):
    return (np.exp(a*x) - np.exp(-a*x))/(np.exp(a*x) + np.exp(-a*x))

def piecewise_linear(x, a=1):
    out = []
    for i, temp in enumerate(x):
        if temp >= 1./a:
            out.append(1)
        elif temp < 1./a and temp > -1./a:
            out.append(a*temp)
        else:
            out.append(-1)
    return np.asarray(out)
```

```python
x = np.linspace(-10,10,100)

def plot_comparison(func, alpha, x, title):
    # Function plots the activations defined above for different values of a and a range x
    colours = ["r", "b", "g", "y"]
    
    for i, a in enumerate(alpha):
        plt.plot(x, func(x, a), colours[i], label='a = {}'.format(a))
    
    plt.title(title)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel(r'$\sigma(x|a)$')
    plt.grid(True)
```

```python
plt.figure(1)
plot_comparison(sigmoid, [0.25, 0.5, 1, 100], x,
                r"Sigmoid Activation Function")
plt.figure(2)
plot_comparison(tanh, [0.25, 0.5, 1, 100], x,
                r"Hyperbolic Tangent Activation Function")
plt.figure(3)
plot_comparison(piecewise_linear, [0.25, 0.5, 1, 100], x,
                r"Piecewise Linear Activation Function")
```

The step function is achieved in the limit of $\alpha \to \infty$ (See e.g. yellow line in plots). 

# Question 3 - Rosenblatt's Perceptron

```python
def gen_data(n, d=2):
    # Function generates data from independent Gaussians
    X = np.empty((n, d+1))
    
    X[:, 0] = - np.ones(n)
    for i in range(1, d+1):
        X[:, i] = np.random.standard_normal(n)
    
    y = (X[:, d] >= 0.5 - X[:, d-1])
    y = map(lambda temp: 1 if temp else -1, y)
    return X, y
```

```python
X_train, y_train = gen_data(1000)
```

```python
class Rosenblatt_Perceptron():
    def __init__(self, X, y, l_rate, max_epoch=100):
        # Initialize data, shapes, learning rate, max_epoch
        self.n, self.d = X.shape
        self.X = X
        self.y = y
        self.l_rate = l_rate
        self.max_epoch = max_epoch
        
        # Initialize weights to 0
        self.weights = np.zeros(self.d)
    
    def update(self, i):
        # Calculate weight update based on error and 
        # return amount of change as vector (for each weight)
        delta = self.l_rate*(self.y[i] - self.predict(self.X[i, :]))*self.X[i, :]
        self.weights += delta
        return delta
    
    def train(self, eps):
        delta = np.infty
        counter = 0
        
        # Run updates over complete data set (one for loop over all obs = 1 epoch)
        # Until norm of weight change vector is below thresh
        while np.linalg.norm(delta) > eps and counter < self.max_epoch:
            for i in range(self.n):
                delta = self.update(i)
            counter += 1
            print("Done training epoch {} - Training Accuracy: {}".format(counter,
                  self.calculate_accuracy(self.X, self.y)))
        return
    
    def predict(self, x):
        # Calculate predicted label for a single observation
        return np.sign(np.dot(self.weights, x))
    
    def calculate_accuracy(self, X, y):
        predictions = []
        # Calculate predicted label for all observation
        # Could be done more efficiently using matrix multiplication and then applying sign
        for i in range(X.shape[0]):
            predictions.append(self.predict(X[i, :]))
        return np.mean(np.equal(y, np.asarray(predictions)))
```

```python
# Initialize the Perceptron with the data and a learning rate
RP = Rosenblatt_Perceptron(X_train, y_train, 0.01, 50)
```

```python
# Perform training until weights converge
RP.train(1e-10)
```

```python
print("Converged weights: {}".format(RP.weights))
```

```python
# Generate Test/Validation Set and calculate accuracy based on above weights
X_test, y_test = gen_data(100)
print("Test Accuracy: {}".format(RP.calculate_accuracy(X_test, y_test)))
```

```python
data_0 = X_train[np.where(np.asarray(y_train) == -1), :]
data_0 = np.reshape(data_0, (data_0.shape[1], data_0.shape[2]))
data_1 = X_train[np.where(np.asarray(y_train) == 1), :]
data_1 = np.reshape(data_1, (data_1.shape[1], data_1.shape[2]))

plt.scatter(data_0[:,1], data_0[:,2], color='r', label="Class 1")
plt.scatter(data_1[:,1], data_1[:,2], color='b', label="Class 2")

a = - RP.weights[2] / RP.weights[1]
xx = np.linspace(-3, 3)
yy = a * xx + (RP.weights[0]) / RP.weights[1]

plt.plot(xx, yy, 'k-', label="Decision Boundary")
plt.legend()

plt.title('Decision Boundary of Rosenblatt Perceptron')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
```

# Question 4 - Linear Separability

```python
def gen_data_xor(n, p=0.5):
    X = np.empty((n, 2))
    y = []
    
    for i in range(n):
        X[i, 0] = np.random.binomial(1, p)
        X[i, 1] = np.random.binomial(1, p)
    
        if X[i, 0] == 0 and X[i, 1] == 0:
            y.append(0)
        elif X[i, 0] == 1 and X[i, 1] == 1: 
            y.append(0)
        else:
            y.append(1)
   
    return X, np.asarray(y)
```

```python
X_train, y_train = gen_data_xor(100)
```

```python
# Initialize the Perceptron with the data and a learning rate
RP = Rosenblatt_Perceptron(X_train, y_train, 0.01, 50)
```

```python
# Perform training until weights converge or 
# for a max of 50 epochs - Here: Does not converge!
RP.train(1e-10)
```

```python
data_0 = X_train[np.where(np.asarray(y_train) == 0), :]
data_0 = np.reshape(data_0, (data_0.shape[1], data_0.shape[2]))
data_1 = X_train[np.where(np.asarray(y_train) == 1), :]
data_1 = np.reshape(data_1, (data_1.shape[1], data_1.shape[2]))

plt.scatter(data_0[:,0], data_0[:,1], color='r', label="Class 1")
plt.scatter(data_1[:,0], data_1[:,1], color='b', label="Class 2")

plt.title('XOR Classification Problem - Decision Boundary of Rosenblatt Perceptron')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.legend()

plt.title('XOR Classification Problem')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
```

There is no linear function that is capable of separating the two classes of the XOR problem. Hence, the problem is not solvable by linear separation. A non-linearity is needed. More specifically, a separating hyperplane in N-dimensions is only capable to shatter N+1 points (VC dimension theory). For our case this means that in N=2 only 3 points can be classified in all possible ways with the perceptron defined above.
