# Question 1: Visual Stimulus

```python
def visual_field(x_0, y_0, delta_x, delta_y):
    """
    Inputs: Initial visual field size in degrees, Sampling rates
    Output: Visual field and corresponding vectors
    """
    x = []
    y = []

    for i in range(int(2*x_0/delta_x) + 1):
        x.append(-x_0 + i*delta_x)
    for j in range(int(2*y_0/delta_y) + 1):
        y.append(-y_0 + j*delta_y)

    x = np.array(x)
    y = np.array(y)
    vf, _ = np.meshgrid(x, y)
    
    return x, y, vf
```

```python
# Plot the visual field
delta_x, delta_y = 0.5, 0.5
x_0, y_0 = 5, 5
x, y, vf = visual_field(x_0=5, y_0=5, delta_x=0.5, delta_y=0.5)
plt.imshow(vf, cmap="gray")
plt.title(r"The Visual Field: $x_0=5$, $y_0=5$, $\Delta = 0.5$, $\Delta = 0.5$")
plt.xlabel(r"$x_i$")
plt.ylabel(r"$y_i$")
```

```python
def sinusoidal_grating(x, y, degree_K, theta, phi, A):
    """
    Inputs: spatial field degrees, stimulus spatial frequency, 
            stimulus orientation, spatial phase, contrast amlitude
    Output: sinusoidal grated degrees
    """
    spatial_freq =  K(degree_K)
    return A*np.cos(spatial_freq*x*np.cos(theta) + spatial_freq*y*np.sin(theta) - phi)

def calc_sinusoidal_grated_grid(x, y, degree_K, A, theta, phi):
    """
    Inputs: spatial field degrees, stimulus spatial frequency, 
            stimulus orientation, spatial phase, contrast amlitude
    Output: sinusoidal grating for full visual field grid
    """
    grated_grid = np.zeros(vf.shape)

    for i in range(grated_grid.shape[0]):
        for j in range(grated_grid.shape[1]):
            grated_grid[i, j] = sinusoidal_grating(x[i], y[j], degree_K, theta, phi, A)
            
    return grated_grid
```

```python
def K(degree):
    # spatial frequency
    return math.pi/degree
A = 1  # contrast amplitude
theta = 0  # spatial phase
phi = 0  # orientation
```

# Question 2: A Model of the Receptive Field

```python
def gabor_function(x, y, degree_k, psi, sd_x, sd_y):
    """
    Inputs: spatial field degrees, preferred spatial frequency (degree), 
            preferred spatial phase, receptive field size (standard devs)
    Output: gabor "kerneled" output
    """
    spatial_freq =  K(degree_k)
    D = 1./(2*math.pi*sd_x*sd_y)*\
        np.exp(-x**2/(2*sd_x**2) -y**2/(2*sd_y**2))*\
        np.cos(spatial_freq*x - phi)
    return D

def calc_gabor_receptive_field(x, y, degree_k, psi, sd_x, sd_y):
    """
    Inputs: spatial field degrees, preferred spatial frequency (degree), 
            preferred spatial phase, receptive field size (standard devs)
    Output: Full Gabor receptive field
    """
    gabor_grid = np.zeros(vf.shape)
    
    for i in range(gabor_grid.shape[0]):
        for j in range(gabor_grid.shape[1]):
            gabor_grid[i, j] = gabor_function(x[i], y[j],
                                              degree, psi, sd_x, sd_y)
            
    return gabor_grid
```

### Optimal Stimulus Specification for such a Visual Field:
Such a neuron can be found in the primary visual cortex (V1 - see e.g. Wiesel and Hubel (1969)). For the optimal stimulus the following relationship to the receptive field hold:
1. $k = K$ - same stimulus spatial frequency and preferred frequency
2. $\theta = 0$ - spatial orientation aliged
3. $\phi = \Psi$ - same spatial and preferred phase

This receptive field characterizes a simple cell since it responds to the absolute position/direction of a stimulus/line element.

```python
def response(x, y, delta_x, delta_y,
             gabor_function, sinusoidal_grating,
             degree_K, degree_k, theta, phi, A, psi, sd_x, sd_y, type="sino"):
    """
    Inputs: All inputs to grating and gabor function as well as sampling parameters
    Output: response of simple cell
    """
    r = 0
    for i in range(len(x)):
        for j in range(len(y)):
            if type == "sino":
                r += sinusoidal_grating(x[i], y[j], degree_K, theta, phi, A)\
                *gabor_function(x[i], y[j], degree_k, psi, sd_x, sd_y)\
                *delta_x*delta_y
            else:
                r += A\
                *gabor_function(x[i], y[j], degree_k, psi, sd_x, sd_y)\
                *delta_x*delta_y
    return r
```

# Question 3: Tuning Curve

```python
orientation_range = np.linspace(0, 180, 90)
K_k_ratio_range = np.linspace(0.1, 5, 90)
phase_range = np.linspace(0, 15, 180)
```

```python
responses_orientation = response(x, y, delta_x, delta_y,
                                 gabor_function, sinusoidal_grating,
                                 degree_K=5, degree_k=5, 
                                 theta=orientation_range, phi=0, 
                                 A=1, psi=0, sd_x=2, sd_y=1) 

responses_K_k_ratio = response(x, y, delta_x, delta_y,
                               gabor_function, sinusoidal_grating,
                               degree_K=1, degree_k=K_k_ratio_range,
                               theta=orientation_range, phi=0, 
                               A=1, psi=0, sd_x=2, sd_y=1) 

responses_phase = response(x, y, delta_x, delta_y,
                           gabor_function, sinusoidal_grating,
                           degree_K=5, degree_k=5, theta=0, 
                           phi=phase_range, A=1, psi=0, sd_x=2, sd_y=1) 
```

```python
plt.plot(orientation_range, responses_orientation)
plt.xlabel(r"$\Theta$")
plt.ylabel(r"Response $r$")
plt.title("Tuning Curve: Stimulus Orientation")
```

```python
plt.plot(K_k_ratio_range, responses_K_k_ratio)
plt.xlabel(r"$K/k$")
plt.ylabel(r"Response $r$")
plt.title("Tuning Curve: Ratio K/k")
```

```python
plt.plot(phase_range, responses_phase)
plt.xlabel(r"$\Phi$")
plt.ylabel(r"Response $r$")
plt.title("Tuning Curve: Spatial Phase")
```

### Receptive Field Characterization
* Stimulus orientation has cyclical effect (ceteris paribus) on response.
* Ratio of stimulus and preferred frequency has an alternating effect on cell response between 0.75 and 2 and afterwards a constant effect.
* Spatial phase has cyclical effect (ceteris paribus) on response.

# Question 4: Image Filtering

```python
img_cat = plt.imread("cat.jpg")

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img_cat_gray = rgb2gray(img_cat)
img_cat_ds = img_cat_gray[::6, ::6]
```

```python
plt.imshow(img_cat_ds, cmap="gray")
plt.title("The Downsampled Image of Choice")
```

```python
def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 255)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value
    return vector
```

```python
def filter(x_0, y_0, delta_x, delta_y,
           image, gabor_function, 
           degree_k, theta, phi, A, psi, sd_x, sd_y):
    
    x, y, vf = visual_field(x_0, y_0, delta_x, delta_y)
    r = np.zeros((image.shape[0], image.shape[1]))
    image_pad = np.pad(image, int(x_0/delta_x), pad_with)
    
    x_k = np.linspace(-5, 5, image.shape[0]) 
    y_l = np.linspace(-5, 5, image.shape[1])
    
    for k in range(image.shape[0]):
        for l in range(image.shape[1]):
            stim = image_pad[k:k+2*int(x_0/delta_x)+1,
                             l:l+2*int(y_0/delta_y)+1]
            for i in range(len(x)):
                for j in range(len(y)):
                    r[k, l] += stim[i, j]\
                    *gabor_function(x[i]-x_k[k], y[j]-y_l[l],
                                    degree_k, psi, sd_x, sd_y)\
                    *delta_x*delta_y
    return r
```

```python
filter_img = filter(x_0=5, y_0=5, delta_x=0.5, delta_y=0.5,
                    image=img_cat_ds, gabor_function=gabor_function, 
                    degree_k=5./8, theta=0, phi=0, A=1,
                    psi=math.pi/2, sd_x=1./8, sd_y=1./8)
```

```python
plt.imshow(filter_img, cmap="gray")
plt.title("The Filtered Image of Choice")
```

Note that we can observe the rough shape of the cat in the filter image. Filter acts as a rough edge/line/shape detector.
